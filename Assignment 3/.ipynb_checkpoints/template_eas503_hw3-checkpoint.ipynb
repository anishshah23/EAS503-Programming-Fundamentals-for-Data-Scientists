### Problem 1 - Chicago Crime Data (25 points)

The Chicago Crime Data reflects the crime incidents that occured in the city of Chicago during 2016 and 2017. The data is provided in the `chicago_crime_data_v2.csv` file. For each crime incident, there is information regarding the case ID, crime location, description, primary crime type ID, coordinates of the crime etc. To identify the primary crime type (denoted by the `Primary type` variable), the `primary_type_ID.csv` file is provided which maps the primary crime type to their corresponding ID.

For this problem, the goal is to identify the crime types that have occurred more than 10,000 times during the years 2016 and 2017. For this,

1. Import the CSV files as dataframes using pandas.

2. Sort the crime data file by crime ID and merge the two files by `Primary Type ID` variable. 

3. Identify the crime types that have a frequency >10,000 and list them as `High Frequency Crime Types`. Provide a timeseries plot for all `High Frequency Crime Types`. For this, compute the number of crimes that occur per each crime type per month (2016 Jan - 2017 Dec). Use the `series.plot` method in Pandas for Series objects. 

4. Using the code provided below, plot all the crimes that come under `High Frequency Crime Types` on a map of Chicago. For this, use the location information provided in the data. For any crimes with missing location coordinates, sort the data by ascending `Date` and use forward fill.  

(Make sure that each data point must be plotted as a transparent circular dots with a markersize of atleast 20 and alpha set to atmost 0.5 (transparency))


```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import importlib
mpl_toolkits = importlib.import_module('mpl_toolkits')
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=[16,10])
m = Basemap(projection='merc',llcrnrlat=41.60,urcrnrlat=42.10,\
        llcrnrlon=-88.0,urcrnrlon=-87.50,lat_ts=20,resolution='c')
m.readshapefile('geo_export_0e3da441-8fe8-4e11-9ca0-42ef75cab68e','chicago')
dataFramePrimaryId = pd.read_csv('primary_type_id.csv')
# print(dataFramePrimaryId["Primary Type ID"][0])
# print(dataFramePrimaryId.sort_values("Primary Type ID", ascending=False)["Primary Type ID"][0])
dataFrameCrime = pd.read_csv('chicago_crime_data.csv')
# print(dataFrameCrime["Crime_ID"])
dataFrameCrime["Crime_ID"] = dataFrameCrime["Crime_ID"].astype('int64')
newFrame = dataFrameCrime.sort_values("Crime_ID", ascending=True)

newFrame = newFrame.reset_index(drop=True)
# print(newFrame["Crime_ID"][0])
mergedFrame = pd.merge(newFrame, dataFramePrimaryId, on='Primary Type ID', how='inner')
posFrame = mergedFrame
posFrame = posFrame.sort_values("Date", ascending=True)
coloumns = ['Latitude', 'Longitude']
posFrame[coloumns] = posFrame[coloumns].ffill()



# print(mergedFrame["Primary Type"])
highFrequencyCrimes = {}
CrimesFrequency = {}
for index, row in mergedFrame.iterrows():
    if row["Primary Type"] in CrimesFrequency:
        CrimesFrequency[row["Primary Type"]].append(row["Date"])
        if len(CrimesFrequency[row["Primary Type"]]) > 10000:
            highFrequencyCrimes[row["Primary Type"]] = CrimesFrequency[row["Primary Type"]]
    else:
        CrimesFrequency[row["Primary Type"]] = []
        CrimesFrequency[row["Primary Type"]].append(row["Date"])

Longitude = []
Latitude = []
for index, row in posFrame.iterrows():
    if row["Primary Type"] in CrimesFrequency:
        CrimesFrequency[row["Primary Type"]].append(row["Date"])
        if len(CrimesFrequency[row["Primary Type"]]) > 10000:
            Longitude.append(row["Longitude"])
            Latitude.append(row["Latitude"])
    else:
        CrimesFrequency[row["Primary Type"]] = []
        CrimesFrequency[row["Primary Type"]].append(row["Date"])

print(Longitude)
print(Latitude)
x,y = m(Longitude,Latitude)
fig = plt.figure(1)
plt.scatter(x,y,alpha=0.5,s=20)
plt.title("Chicago Cenus Tracts")
# plt.show()

# for key, value in highFrequencyCrimes.items():
#     print(key)

print("--------------------------------------------------------------")
# for key, value in CrimesFrequency.items():
#     print(key)
# print(CrimesFrequency)
# highFrequencyCrimes = {"ASSAULT" : ["1/12/2016 19:30", "1/13/2016 19:30", "1/12/2016 19:30", "2/13/2016 19:30", "2/12/2016 19:30", "2/13/2016 19:30"]}
parseGraph = {}
for key, value in highFrequencyCrimes.items():
    parseGraph[key] = {}
    for date in value:
        dateObject = datetime.strptime(date, '%m/%d/%Y %H:%M')
        name = dateObject.strftime('%b')
        year = dateObject.year
        if (name+str(year)) in parseGraph[key]:
            parseGraph[key][(name+str(year))] = parseGraph[key][(name+str(year))] + 1
        else:
            parseGraph[key][(name+str(year))] = 1

# objPython = pd.DataFrame.from_dict(data=parseGraph, orient='columns', dtype=None)
print(parseGraph["ASSAULT"])
index = ['Jan2016', 'Feb2016', 'Mar2016', 'Apr2016', 'May2016', 'Jun2016',
         'Jul2016', 'Aug2016', 'Sep2016', 'Oct2016', 'Nov2016', 'Dec2016',
         'Jan2017', 'Feb2017', 'Mar2017', 'Apr2017', 'May2017', 'Jun2017',
         'Jul2017', 'Aug2017', 'Sep2017', 'Oct2017', 'Nov2017', 'Dec2017']

fig2 = plt.figure(2,figsize=(30, 10))
for key, value in parseGraph.items():
  seriesObject = pd.Series(parseGraph[key], index=index)
  seriesObject.index = pd.CategoricalIndex(seriesObject.index,
                               categories=index,
                               sorted=True)
  seriesObject  = seriesObject.sort_index()

  ax = seriesObject.plot(label=key, legend=True)
  ax.set_xticks(range(len(index)));
  ax.set_xticklabels(index, rotation=90)


plt.show()
print(seriesObject)
```
